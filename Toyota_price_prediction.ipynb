As our dataset has no missing values and it is clean, We can move further to implement pycaret library for predicting prices based on trained data

```python
from pycaret.regression import *
s=setup(df,target='Price',session_id=123)
```

Compare_models() compares all the regression models and selects the best model based on MSE,MAE,RMSE,R2, and Time taken to execute

```python
model=compare_models()
```

We split the data into features and target for further implementation 

```python
from sklearn.model_selection import train_test_split

# Splitting the data into features and target
features = df[['Weight', 'Doors', 'HP', 'MetColor', 'CC', 'FuelType', 'KM', 'Age', 'Automatic']]
target = df['Price']  # Extract 'Price' column from the DataFrame

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
print("Training set size: ", len
      (y_train))  # The number
# of samples in the training set is
# approximately equal to 75% of all
# available samples
print("Testing set size: ", len(
    y_test), "\n")  
```

Predict_model is used to predict the price based on the model we pass and test data

```python
predictions = predict_model(model, data=X_test)
predictions
```
